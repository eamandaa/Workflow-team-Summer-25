# Generate empty WGS of raw, processed and summarised files for 300 patients with 1 sample each. 

## Import libraries

```python
import os
```

```python
num_samples = 300
base_path = "WGS"

# Create subfolders
raw_folder = os.path.join(base_path, "Raw")
processed_folder = os.path.join(base_path, "Processed")
summ_folder = os.path.join(base_path, "Summarised")

os.makedirs(raw_folder, exist_ok=True)
os.makedirs(processed_folder, exist_ok=True)
os.makedirs(summ_folder, exist_ok=True)

for i in range(1, num_samples + 1):
    sample_id = f"XY{i:03d}-AGRF"
    
    # File paths
    raw_file = os.path.join(raw_folder, f"{sample_id}.fastq")
    processed_file = os.path.join(processed_folder, f"{sample_id}.bam")
    summary_file = os.path.join(summ_folder, f"{sample_id}.vcf")
    
    # Create empty files
    open(raw_file, "w").close()
    open(processed_file, "w").close()
    open(summary_file, "w").close()
```
