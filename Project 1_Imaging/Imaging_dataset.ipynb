# Generate empty WGS of raw, processed and summarised files for 300 patients with 1 sample each. 

## Import libraries

```python
import os
```

```python
num_patients = 300

for i in range(1, num_patients + 1):
    # Create patient + sample IDs
    patient_id = f"Patient{i:03d}_DiseaseX"
    sample_id = f"XY{i:03d}"   # For filenames
    sample_folder_name = f"Sample{sample_id}"  # Folder name
    
    # Folder structure
    patient_folder = patient_id
    sample_folder = os.path.join(patient_folder, sample_folder_name)
    
    # Create directories
    os.makedirs(sample_folder, exist_ok=True)
    
    # File names with AGRF
    file_paths = [
        os.path.join(sample_folder, f"{sample_id}-AGRF.czi"),
        os.path.join(sample_folder, f"{sample_id}-AGRF.tif"),
        os.path.join(sample_folder, f"{sample_id}-AGRF.zarr")
    ]
    
    # Create empty files
    for file_path in file_paths:
        open(file_path, 'w').close()
```
